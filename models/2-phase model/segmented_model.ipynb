# With SMOTE

## Start

```python
from src.segmented_model import *
import pandas as pd

X_train_unb = pd.read_csv('../../data/processed/interim/X_train_X_train_unbalanced.csv')
y_train_unb = pd.read_csv('../../data/processed/y_train_p.csv')

X_val = pd.read_csv('../../data/processed/X_val_p.csv')
y_val = pd.read_csv('../../data/processed/y_val.csv')

X_train, y_train = randomOverSample(X_train_unb, y_train_unb, random_state=42)
```

## Phase 1: Segmentation

### K-means Segmentation

#### 2-cluster K-means

```python
km_clusters = 2
df_kmeans_segmented2 = kmeans_segmentation(km_clusters, X_train, y_train)
visualize('kmeans_segment', df_kmeans_segmented2, km_clusters) 
```

#### 3-cluster K-means

```python
km_clusters = 3
df_kmeans_segmented3 = kmeans_segmentation(km_clusters, X_train, y_train)
visualize('kmeans_segment', df_kmeans_segmented3, km_clusters) 
```

#### 4-cluster K-means

```python
km_clusters = 4
df_kmeans_segmented4 = kmeans_segmentation(km_clusters, X_train, y_train)
visualize('kmeans_segment', df_kmeans_segmented4, km_clusters) 
```

#### 5-cluster K-means

```python
km_clusters = 5
df_kmeans_segmented5 = kmeans_segmentation(km_clusters, X_train, y_train)
visualize('kmeans_segment', df_kmeans_segmented5, km_clusters) 
```

#### 6-cluster K-means

```python
km_clusters = 6
df_kmeans_segmented6 = kmeans_segmentation(km_clusters, X_train, y_train)
visualize('kmeans_segment', df_kmeans_segmented6, km_clusters) 
```

### Decision Trees Segmentation

#### Decision trees, max_leafs = 2


```python
dt_leafs = 2
df_dt_segmented2 = decision_trees_segmentation(dt_leafs, X_train, y_train)
visualize('dt_segment', df_dt_segmented2, dt_leafs)
```

#### Decision trees, max_leafs = 3

```python
dt_leafs = 3
df_dt_segmented3 = decision_trees_segmentation(dt_leafs, X_train, y_train)
visualize('dt_segment', df_dt_segmented3, dt_leafs) 
```

#### Decision trees, max_leafs = 4

```python
dt_leafs = 4
df_dt_segmented4 = decision_trees_segmentation(dt_leafs, X_train, y_train)
visualize('dt_segment', df_dt_segmented4, dt_leafs) 
```

#### Decision trees, max_leafs = 5

```python
dt_leafs = 5
df_dt_segmented5 = decision_trees_segmentation(dt_leafs, X_train, y_train)
visualize('dt_segment', df_dt_segmented5, dt_leafs) 
```

#### Decision trees, max_leafs = 6

```python
dt_leafs = 6
df_dt_segmented6 = decision_trees_segmentation(dt_leafs, X_train, y_train)
visualize('dt_segment', df_dt_segmented6, dt_leafs) 
```

## ---------------

## Phase 2: Logistic Regression

```python
segmentations = [df_kmeans_segmented2,
                 df_kmeans_segmented3,
                 df_kmeans_segmented4,
                 df_kmeans_segmented5,
                 df_kmeans_segmented6,
                 df_dt_segmented2,
                 df_dt_segmented3,
                 df_dt_segmented4,
                 df_dt_segmented5,
                 df_dt_segmented6]

segmentation_names=['2-cluster K-means',
                    '3-cluster K-means',
                    '4-cluster K-means',
                    '5-cluster K-means',
                    '6-cluster K-means',
                    '2-cluster Decision Trees',                    
                    '3-cluster Decision Trees',
                    '4-cluster Decision Trees',
                    '5-cluster Decision Trees',
                    '6-cluster Decision Trees']
```

```python
all_metrics = []
for i in range(len(segmentations)):
    print("-"*50,"\n-----", segmentation_names[i], "-----\n")
    case_metrics = second_phase(segmentations[i], X_val, y_val)
    all_metrics.append(case_metrics)
```

```python
print("Metrics Summary", "\n(Mean value for each metric)")
print(metrics_summary(all_metrics, segmentation_names))
```
