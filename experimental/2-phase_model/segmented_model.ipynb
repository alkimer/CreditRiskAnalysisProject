# ---  2-phase Model ---

# Introduction

This two-phase model was designed as an experimental approach to tackle the problem from multiple angles and explore alternative solutions with the expectation of improving the performance metrics.

>> Given that the original dataset is imbalnced with respect to the target variable (74% for class '0', non-risk, and 26% for class '1', credirisk), two versions of this model were tried.

>> In the first section we worked with balanced data with SMOTE (Synthetic Minority Oversampling Technique). In the second section we used the imbalanced data with SMOTE in the second section.

In this case, not applying oversampling through SMOTE resulted in a dramatic fall in performance metrics, as you can see in the 'Results' section, at the end.

The best metrics achieved by this model did not meet a a satisfactory standard, likely due to poor quality data.

# 4. Conclusions

> 1. **Impact of Skipping SMOTE on Performance**.
> In this scenario, skipping SMOTE-based oversampling led to a dramatic decline in performance metrics, as shown in the final 'Results' section.

> 2. **Effect of Data Quality on Model Outcomes**.
> Even the models best-performing metrics fell short of acceptable standards, likely as a consequence of low-quality data.

> 3. **The importance of data balancing**.
> Applying techniques like SMOTE to balance classes is crucial for improving model performance (at least for this model), especially when the original dataset is imbalanced.

> 4. **Limitations due to data quality**.
> Poor quality or unrepresentative data significantly affects the models ability to generalize, suggesting the need to improve data collection or preprocessing.

> 5. **Segmentation as a strategy**.
> Using segmentation (e.g., K-means clustering) can help address data heterogeneity, but its effectiveness depends on the quality of segmentation, quality of data and the size of each segment.

> 6. **Continuous experimentation and evaluation**.
> Experimenting with different configurations and models is necessary to find the best solution, especially in complex problems where no single model fits all cases.

> 7. **Next Steps: Exploring Alternative Models**.
>Given the results above, we proceeded to explore other models hoping achieving better performance for the same objective.
